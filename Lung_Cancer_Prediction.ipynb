```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Srinivasa Rao/OneDrive/Documents/GUVI Files/Datasets/lung cancer survey.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['GENDER'] = le.fit_transform(df['GENDER'])
df['LUNG_CANCER'] = le.fit_transform(df['LUNG_CANCER'])
print(df)
```

```python
corrm = df.corr()
corrm
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(corrm, annot=True)
```

```python
df.columns = df.columns.str.strip() 
```

```python
print(df.columns)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  accuracy_score, RocCurveDisplay, f1_score,roc_auc_score,recall_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df[['GENDER','AGE','SMOKING','YELLOW_FINGERS','ANXIETY','PEER_PRESSURE','CHRONIC DISEASE','FATIGUE','ALLERGY','WHEEZING','ALCOHOL CONSUMING','COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY','CHEST PAIN']]
y = df[['LUNG_CANCER']]

scaler_classify = StandardScaler().fit(X)
X_scaled = scaler_classify.transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.27, random_state=27)
```

```python
import pickle

file_path='scaling_classify.pkl'
with open(file_path, 'wb') as file:
    pickle.dump(scaler_classify, file)
print(f'Pickle file created: {file_path}')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid of plots

# Model list and axes
models = [
    ("Random Forest", RandomForestClassifier(), axes[0, 0]),
    ("Decision Tree", DecisionTreeClassifier(), axes[0, 1]),
    ("XGB Classifier", XGBClassifier(), axes[1, 0]),
    ("KNN Classifier", KNeighborsClassifier(), axes[1, 1])
]

# Loop through models and fit/predict/score each
for name, model, ax in models:
    # Fit the model
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]  # Probabilities for ROC

    # Compute metrics
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    # Plot ROC curve
    RocCurveDisplay.from_predictions(y_test, y_prob, ax=ax)
    ax.set_title(f"{name}\nAccuracy: {accuracy:.2f}, F1: {f1:.2f}, Recall: {recall:.2f}, AUC: {auc:.2f}")

# Show the ROC curves
plt.tight_layout()
plt.show()
```
